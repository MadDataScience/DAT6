```python
import pandas as pd
import yaml
```

```python
## Function to get data - maybe add logic to parse
def fetchTDETemplateData(template_id, usr, pwd, report_start, report_end):
    # returns a data frame
    url = "https://api-preview.telogis.com/execute?template=%s&user=%s&pass=%s&ReportStart=%s&ReportEnd=%s" %(template_id, usr, pwd, report_start, report_end)
    #print url
    df = pd.read_csv(url, header=0)
    return df
```
