#### Setup

```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import sqlite3
from scipy import stats
import scipy.stats
from pandas.io import sql
from scipy.stats import t as tdist
%load_ext sql
conn = sqlite3.connect('enron.db')
```

#### Select a DataFrame containing the recipient count (as before) and the department of the sender for each message.

```python
# qry into a DataFrame
mrc = sql.frame_query("""SELECT mid, max(rno) as to_count, department
                        FROM MessageBase
                        JOIN RecipientBase Using(mid)
                        JOIN EmployeeBase ON from_eid = eid
                        GROUP BY mid""", 
                      conn, "mid")
#mrc.head()
```

####Create a histogram of the recipient count.

```python
mrc.to_count.hist(bins=range(60))
```

####Create a new column based on the log of the recipient count.

```python
mrc['ln_toCount'] = np.log(mrc.to_count)
#mrc.ln_toCount.head()
```

####Create a histogram of that log.

```python
mrc.ln_toCount.hist()
```

####Create a boxplot of the log, splitting the data based on the department of the sender.

```python
mrc.boxplot(column="ln_toCount", by="department", figsize=(12, 4), vert=False, widths=.75)
```

#### Compute the sample mean and standard deviation of the log in the three groups.

```python
_Trading = mrc.ln_toCount[mrc.department == 'Trading']
print ('Trading\n\tavg: %f\n\tstd: %f') %(_Trading.mean(), _Trading.std())
_Other = mrc.ln_toCount[mrc.department == 'Other']
print ('Other\n\tavg: %f\n\tstd: %f') %(_Other.mean(), _Other.std())
_Legal = mrc.ln_toCount[mrc.department == 'Legal']
print ('Legal\n\tavg: %f\n\tstd: %f') %(_Legal.mean(), _Legal.std())
```

####Compute a 95% confidence interval for the difference in recipient count between the three groups.

```python
def CI(sample, CI):
    N = sample.count()
    cutoff = tdist.isf(CI,N-1)
    return(sample.mean() - cutoff*sample.std()/np.sqrt(N),
           sample.mean() + cutoff*sample.std()/np.sqrt(N))

print "\t95% Confidence Interval"
_Trading = mrc.to_count[mrc.department == 'Trading']
l,u = CI(_Trading, 0.95)
print "Trading\n\tL: {}\tU: {}".format(l,u)
_Other = mrc.to_count[mrc.department == 'Other']
l,u = CI(_Other, 0.95)
print "Other\n\tL: {}\tU: {}".format(l,u)
_Legal = mrc.to_count[mrc.department == 'Legal']
l,u = CI(_Legal, 0.95)
print "Legal\n\tL: {}\t\tU: {}".format(l,u)
```

####At level α=5%, test the null hypothesis that the average recipient count does not differ between the three groups. What assumptions are you making? What can you conclude?

```python
mod = sm.OLS.from_formula("to_count ~ department", mrc)
res = mod.fit()
print res.summary()
```
