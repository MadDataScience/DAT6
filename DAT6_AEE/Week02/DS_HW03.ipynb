```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
from pandas.io import sql
import matplotlib.pyplot as plt
conn = sqlite3.connect('enron.db')
```

```python
SubjectLength = sql.frame_query("""SELECT mid, datetime(unix_time,'unixepoch') as time_sent, length(subject) as len_sub
                                    FROM MessageBase""", conn, "mid")
SubjectLength.head()
```

```python
% matplotlib inline
w = SubjectLength.time_sent.map(lambda dt: pd.Period(dt, freq='W'))
SubjectLength.groupby(w).len_sub.median().plot()
plt.xlabel('Date (grouped weekly)')
plt.ylabel('median subject length')
```
