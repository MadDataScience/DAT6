```python
%%sql sqlite:///shopping.db
-- Number of shoppers in transactions0

SELECT COUNT(DISTINCT id) AS num_ids FROM transactions0
```

```python
%%sql sqlite:///shopping.db
-- Number of shoppers in transactions0 and history

SELECT COUNT(t.id) AS num_ids
FROM
    (
    SELECT DISTINCT id FROM transactions0
    ) t
INNER JOIN
    (
    SELECT DISTINCT id FROM history
    ) h
ON t.id = h.id
```

```python
%%sql sqlite:///shopping.db
-- Number of shoppers in history

SELECT COUNT(DISTINCT transactions0.id) AS num_ids
FROM transactions0
INNER JOIN history
ON transactions0.id = history.id
```

```python
%%sql sqlite:///shopping.db
-- Total number of purchases each shopper made

SELECT transactions0.id, repeater, COUNT(1) AS num_purchases
FROM transactions0
INNER JOIN history
ON transactions0.id = history.id
GROUP BY 1, 2
LIMIT 10
```

```python
%%sql sqlite:///shopping.db
-- Total number of shopping trips each shopper made

SELECT id, COUNT(DISTINCT date) AS num_trips
FROM transactions0
GROUP BY 1
LIMIT 10
```

```python
%%sql sqlite:///shopping.db
-- Possible values for offer value

SELECT DISTINCT offervalue
FROM offers
ORDER BY 1
```

```python
%%sql sqlite:///shopping.db
-- Looking at the history-offers joint table

SELECT * FROM history_offers LIMIT 10
```

```python
%%sql sqlite:///shopping.db
-- Delay between last shopping trip and date of offer

SELECT id, (strftime('%s',ho_date) - strftime('%s',trans_date))/(3600*24) AS delay_days
FROM
    (
    SELECT trans.id, MAX(trans.date) AS trans_date, MAX(ho.date) AS ho_date
    FROM
        (
        SELECT DISTINCT id, date
        FROM transactions0
        ) trans
    INNER JOIN
        (
        SELECT DISTINCT id, offerdate AS date
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.date <= ho.date
    GROUP BY 1
    ) m
LIMIT 10
```

```python
%%sql sqlite:///shopping.db
-- Number of chains each shopper shopped at:  Turns out it's always 1
    
SELECT num_chains, COUNT(DISTINCT id) AS num_ids
FROM
    (
    SELECT id, COUNT (DISTINCT chain) AS num_chains
    FROM transactions0
    GROUP BY 1
    ) m
GROUP BY 1
ORDER BY 1
```

```python
%%sql sqlite:///shopping.db
-- Number of shoppers for whom the offer chain was a chain they had previously shopped at

SELECT
    CASE WHEN id_offer IS NOT NULL THEN 1 ELSE 0 END AS chain_match
    , COUNT(DISTINCT id_trans) AS num_ids
FROM
    (
    SELECT trans.id AS id_trans, ho.id AS id_offer
    FROM
        (
        SELECT DISTINCT id, chain
        FROM transactions0
        ) trans
    LEFT JOIN
        (
        SELECT DISTINCT id, chain
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.chain = ho.chain
    ) m
GROUP BY 1
ORDER BY 1 DESC
```

```python
%%sql sqlite:///shopping.db
-- Number of times shopper had purchased the brand of offer before

SELECT id
    , MAX(CASE WHEN id_offer IS NOT NULL THEN num_purchases ELSE 0 END) AS num_match_brand
FROM
    (
    SELECT trans.id, ho.id AS id_offer, num_purchases
    FROM
        (
        SELECT id, brand, COUNT(1) AS num_purchases
        FROM transactions0
        ) trans
    LEFT JOIN
        (
        SELECT DISTINCT id, brand
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.brand = ho.brand
    ) m
GROUP BY 1
LIMIT 10
```

#### Of 5431 shoppers (in transactions0 subset), 3152 received an offer from a chain they had purchased from, and 2279 from one they had not.

```python
%%sql sqlite:///shopping.db
-- Number of shoppers for whom the offer category was a category they had previously purchased

SELECT
    CASE WHEN id_offer IS NOT NULL THEN 1 ELSE 0 END AS category_match
    , COUNT(DISTINCT id_trans) AS num_ids
FROM
    (
    SELECT trans.id AS id_trans, ho.id AS id_offer
    FROM
        (
        SELECT DISTINCT id, category
        FROM transactions0
        ) trans
    LEFT JOIN
        (
        SELECT DISTINCT id, category
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.category = ho.category
    ) m
GROUP BY 1
ORDER BY 1 DESC
```

```python
%%sql sqlite:///shopping.db
-- Number of times each shopper had purchased the offer category before

SELECT id_trans
    , MAX(CASE WHEN id_offer IS NOT NULL THEN num_purchases ELSE 0 END) AS num_purchases_category
FROM
    (
    SELECT trans.id AS id_trans, ho.id AS id_offer, num_purchases
    FROM
        (
        SELECT id, category, COUNT(1) AS num_purchases
        FROM transactions0
        GROUP BY 1, 2
        ) trans
    LEFT JOIN
        (
        SELECT DISTINCT id, category
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.category = ho.category
    ) m
GROUP BY 1
LIMIT 100
```

```python
%%sql sqlite:///shopping.db
-- Total number of distinct categories

SELECT COUNT(DISTINCT category) AS num_cats FROM transactions0
```

```python
%%sql sqlite:///shopping.db
-- Total number of distinct brands

SELECT COUNT(DISTINCT brand) AS num_brands FROM transactions0
```

```python
%%sql sqlite:///shopping.db
-- Number of shoppers for whom the offer brand was a brand they had previously purchased

SELECT
    CASE WHEN id_offer IS NOT NULL THEN 1 ELSE 0 END AS brand_match
    , COUNT(DISTINCT id_trans) AS num_ids
FROM
    (
    SELECT trans.id AS id_trans, ho.id AS id_offer
    FROM
        (
        SELECT DISTINCT id, brand
        FROM transactions0
        ) trans
    LEFT JOIN
        (
        SELECT DISTINCT id, brand
        FROM history_offers
        ) ho
    ON trans.id = ho.id
    AND trans.brand = ho.brand
    ) m
GROUP BY 1
ORDER BY 1 DESC
```
