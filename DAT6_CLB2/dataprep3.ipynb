### Add more shoppers to variables table

```python
%%sql sqlite:///shopping.db
SELECT sh.id
    , (CASE WHEN tr.id IS NOT NULL THEN 5 ELSE table_num END) AS table_num
    , repeater, repeattrips, offerdate, offer, chain, market
    , category, quantity, company, offervalue, brand
    , (CASE WHEN tr.id IS NOT NULL THEN tr.new_num_purchases
           ELSE num_purchases END) AS num_purchases
    , (CASE WHEN tr.id IS NOT NULL THEN tr.new_num_trips
           ELSE num_trips END) AS num_trips    
    , min_date, max_date, offer_delay
    , shopper_age, shopper_lifetime
    , total_purs_spend, num_match_cat, num_match_brand
FROM shopvars sh
LEFT JOIN
    (
    SELECT transactions5.id
        , COUNT(1) AS new_num_purchases
        , COUNT(DISTINCT date) AS new_num_trips
    FROM transactions5
    INNER JOIN shopvars
    ON transactions5.id = shopvars.id
    GROUP BY 1
    ) tr
ON sh.id = tr.id
LIMIT 3
```

```python
%%sql sqlite:///shopping.db
CREATE TABLE shopview AS
SELECT sh.id
    , (CASE WHEN tr.id IS NOT NULL THEN 5 ELSE table_num END) AS table_num
    , repeater, repeattrips, offerdate, offer, chain, market
    , category, quantity, company, offervalue, brand
    , (CASE WHEN tr.id IS NOT NULL THEN tr.new_num_purchases
           ELSE num_purchases END) AS num_purchases
    , (CASE WHEN tr.id IS NOT NULL THEN tr.new_num_trips
           ELSE num_trips END) AS num_trips    
    , min_date, max_date, offer_delay
    , shopper_age, shopper_lifetime
    , total_purs_spend, num_match_cat, num_match_brand
FROM shopvars sh
LEFT JOIN
    (
    SELECT transactions5.id
        , COUNT(1) AS new_num_purchases
        , COUNT(DISTINCT date) AS new_num_trips
    FROM transactions5
    INNER JOIN shopvars
    ON transactions5.id = shopvars.id
    GROUP BY 1
    ) tr
ON sh.id = tr.id
```
