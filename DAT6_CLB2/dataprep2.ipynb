```python
%%sql sqlite:///shopping.db
CREATE TABLE cat_dist AS
SELECT category, num_ids, num_ids_rep, num_ids_non, ROUND(1.0*num_ids_rep/num_ids,5.0) AS pct_rep
    , ABS(0.5 - ROUND(1.0*num_ids_rep/num_ids,3.0)) AS polar_score
FROM
    (
    SELECT category
        , COUNT(DISTINCT tr.id) AS num_ids
        , COUNT(DISTINCT CASE WHEN repeater = 1 THEN tr.id ELSE NULL END) AS num_ids_rep
        , COUNT(DISTINCT CASE WHEN repeater = 0 THEN tr.id ELSE NULL END) AS num_ids_non
    FROM transactions0 tr
    INNER JOIN history h
    ON tr.id = h.id
    GROUP BY 1
    ) m
ORDER BY 6 DESC
```
